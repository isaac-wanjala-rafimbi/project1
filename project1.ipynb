```python
import pandas as pd
```

```python
df=pd.read_csv("1. Weather Data.csv")
```

```python
weather=df[df["Weather"]=="Clear"]
print(weather)
```

```python
wind_speed_4_kmh_count = df[df['Wind Speed_km/h'] == 4].shape[0]
print(wind_speed_4_kmh_count)
```

```python
# Check if there are any NULL values present in the dataset
null_values = df.isnull().sum().sum()
print(null_values)
```

```python
# Rename the column "Weather" to "Weather_Condition"
df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
```

```python
# What is the mean visibility of the dataset?
mean_visibility = df['Visibility_km'].mean()
print(mean_visibility)
```

```python
# Find the number of records where the wind speed is greater than 24 km/hr and visibility is equal to 25 km
wind_speed_gt_24_and_visibility_25_count = df[(df['Wind Speed_km/h'] > 24) & (df['Visibility_km'] == 25)].shape[0]
print(wind_speed_gt_24_and_visibility_25_count)
```

```python
# What is the mean value of each column for each weather condition?
mean_values_per_weather_condition = df.groupby('Weather_Condition').mean()

print(mean_values_per_weather_condition)
```

```python
# Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40
clear_and_humidity_or_visibility = df[(df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50) | (df['Visibility_km'] > 40)]
print(clear_and_humidity_or_visibility)
```

```python
# Find the number of weather conditions that include snow
snow_weather_conditions_count = df[df['Weather_Condition'].str.contains('Snow', na=False)].shape[0]
print(snow_weather_conditions_count)
```
